```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
from resizeimage import resizeimage
```

```python
img = cv2.imread("early_1800.jpg")

averaging = cv2.blur(img, (21, 21))
gaussian = cv2.GaussianBlur(img, (21, 21), 0)
median = cv2.medianBlur(img, 5)
bilateral = cv2.bilateralFilter(img, 9, 350, 350)

```

```python
plt.title("Orignal Image")
plt.imshow(img)
plt.show()
```

```python
plt.title("Average Image")
plt.imshow(averaging)
plt.show()
```

```python
plt.title("Gaussian Image")
plt.imshow(gaussian)
plt.show()
```

```python
plt.title("Median Image")
plt.imshow(median)
plt.show
```

```python
plt.title("BilateralImage")
plt.imshow(bilateral)
plt.show
```

```python
image = cv2.imread("ball.png")
```

```python
plt.title("Orignal Image")
plt.imshow(image)
plt.show()
```

```python
median = cv2.medianBlur(image, 5)
plt.title("Median Image")
plt.imshow(median)
plt.show
```

```python
averaging = cv2.blur(image, (21, 21))
plt.title("Average Image")
plt.imshow(averaging)
plt.show()
```

```python
gaussian = cv2.GaussianBlur(image, (21, 21), 0)
plt.title("Gaussian Image")
plt.imshow(gaussian)
plt.show()
```

```python
bilateral = cv2.bilateralFilter(image, 9, 7.5, 7.5)
plt.title("BilateralImage")
plt.imshow(bilateral)
plt.show
```
